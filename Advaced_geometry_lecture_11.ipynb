- 定義C.12 は　$L$スペクトラムを定義しています。飛躍というのは前回は、形式群を学びました。この形式群が対応するコホモロジーを支配する。そのあたりをK理論のところで見る話とか諸々です。ここでは形式群はAddtiveとMulcative　と　もう一つ　楕円曲線Elliptic Curve の３つになることがわかっており、前者２つが通常のコホモロジーとK理論になってます。そこで楕円曲線からコホモロジーを作りましょうって話で、そこで必要になるのが$E_{\infty}$ スペクトラムなのですが、そのために必要なのが$L$スペクトラムです。スペクトラムはコホモロジー表現するものですが、要は解析で言うと作用を積分で関数をかけて積分するような積分作用素
$g(x)=\int K(x,y)f(y)dy$ の$K(x,y)$ のようなものでしょうか。作用素を色々操作するより　$K(x,y)$ という関数を操作するほうが扱いやすい。この関数がいろんな操作ができると作用素も操作できるといった具合です。今は関数でなく空間（位相空間）ですが、どちらと言うとカテゴリーの言葉で語らていく。もとのスペクトラムとホモトピー同値かホモトピー弱同値で動かして、もっと扱いやすいものに置き換えるその過程で、モナドが出てきたわけです。
